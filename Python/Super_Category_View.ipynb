- User Defined Functions

```python
# Advance Report
def fill_Neo(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[source_df['New Alpha/MP'] == main_col]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# Pipiline report
def fill_Non_Neo(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['New Alpha/MP'] == main_col) & (source_df['Rate Card Bifurcation'] == "Sponsorship")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# Credits
def fill_credits(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[source_df['Alpha_MP'] == main_col]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# Brand_PLA
def fill_brand_pla(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "BRAND_PLA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# Brand_PCA
def fill_brand_pca(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Campaign Type'] == "BRAND_PCA")]
    summary = filtered.groupby(['BU', 'Super category'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# PLA
def fill_pla(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "PLA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# PCA
def fill_pca(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "PCA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# Paid Credits
def fill_paid_credits(template_df, source_df, main_col, sub_col, sum_col):
    summary = source_df.groupby('New Supercat')[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# 'Display'
def fill_display(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['New Alpha/MP'] == main_col) & (source_df['Check'] == "Yes")]
    summary = filtered.groupby('New Supercat')[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

```python
# 'Search'
def fill_search(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[source_df['Alpha_MP'] == main_col]
    summary = filtered.groupby('New Supercategory')[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```
