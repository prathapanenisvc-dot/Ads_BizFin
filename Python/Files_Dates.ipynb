```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog
import os
```

```python

def load_files_with_logic():
    root = tk.Tk()
    root.withdraw()
    root.attributes("-topmost", True)

    # --- CONFIGURATION: Hard-code your special formatting rules here ---
    special_rules = {
        "Advance Report": {"skiprows": 2},
        "Search Billing Sign off": {
            "skiprows": 1, "skipfooter": 9, "engine": 'python'
        },
        "Billing Report": {"sheet_name": "LineItem Report"},
        "Pipeline Report": {"sheet_name": "Line Item Export"}
    }

    data_frames = {}

    while True:
        files = filedialog.askopenfilenames(
            title="Select Files (CSV/XLSX)",
            filetypes=[("Data Files", "*.csv *.xlsx *.xls")]
        )
        
        if not files: break

        for path in files:
            file_name = os.path.basename(path)
            clean_name = os.path.splitext(file_name)[0]
            
            # Check if this filename has special rules defined above
            args = special_rules.get(clean_name, {})

            try:
                if path.endswith('.csv'):
                    # Merge default CSV settings with special rules
                    csv_defaults = {"sep": None, "engine": 'python'}
                    data_frames[clean_name] = pd.read_csv(path, **{**csv_defaults, **args})
                else:
                    # Load Excel with potential sheet_name rules
                    data_frames[clean_name] = pd.read_excel(path, **args)
                
                print(f"✅ Loaded: {clean_name}")
                
            except Exception as e:
                print(f"❌ Error loading {file_name}: {e}")

        if input("Do you want to select files? 🤔(y/n): ").lower() != 'y':
            break

    return data_frames

```

```python
# Search, PC, GMV files
dfs = load_files_with_logic()

#store as dfs for runtime
%store dfs 
```

```python
# Mapping Files
df = load_files_with_logic()
%store df
```
